## Load Plan Mode Template

```python
# Load the plan mode template from repository root
with open("../../PLAN_MODE_TEMPLATE.md") as f:
  PLAN_MODE_TEMPLATE = f.read()

print("Plan Mode Template loaded successfully")
```

## Plan Mode Generator Function

```python
def generate_plan(task_description: str, api_key: str = None) -> str:
  """
  Generate a plan using Claude with Plan Mode Style Guide.

  Args:
      task_description: The task to plan
      api_key: Anthropic API key (or use ANTHROPIC_API_KEY env var)

  Returns:
      Formatted plan following Plan Mode Style Guide
  """
  client = Anthropic(api_key=api_key or os.environ.get("ANTHROPIC_API_KEY"))

  system_prompt = f"""
You are a planning assistant that strictly follows the Plan Mode Style Guide v1.0.

Rules:
- Sacrifice grammar for concision
- One action per line
- Use imperative verbs only
- Keep lines ≤60 chars
- End with 'Unresolved Qs:' section
- Use 'Impl:' if >1 module affected
- No prose explanations

Template:
{PLAN_MODE_TEMPLATE}
"""

  message = client.messages.create(
    model="claude-sonnet-4-5-20250929",
    max_tokens=2048,
    system=system_prompt,
    messages=[{"role": "user", "content": f"Generate a plan for: {task_description}"}],
  )

  return message.content[0].text
```

## Example Usage

```python
# Example: Generate a plan for migrating a database adapter
task = """
Migrate SQLite database implementation to use an adapter pattern
that supports multiple database backends while maintaining backward compatibility.
"""

plan = generate_plan(task)
print(plan)
```

## Batch Plan Generation for GKE Deployment

```python
# Generate plans for multiple deployment tasks
deployment_tasks = [
  "Set up GKE cluster with autoscaling for FastAPI services",
  "Configure Cloud Build CI/CD pipeline for automated deployments",
  "Implement monitoring and alerting with Cloud Operations",
  "Set up secret management with Google Secret Manager",
]

plans = {}
for task in deployment_tasks:
  print(f"\n{'=' * 60}")
  print(f"Task: {task}")
  print("=" * 60)
  plan = generate_plan(task)
  plans[task] = plan
  print(plan)
```

## Save Plans to File

```python
# Save generated plans for documentation
import json
from datetime import datetime

output_file = f"generated_plans_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"

with open(output_file, "w") as f:
  json.dump(plans, f, indent=2)

print(f"Plans saved to: {output_file}")
```
