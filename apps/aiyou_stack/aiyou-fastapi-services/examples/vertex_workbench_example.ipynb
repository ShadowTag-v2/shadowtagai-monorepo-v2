# Code Refactorer Service - Vertex AI Workbench Example

This notebook demonstrates how to use the AI Code Refactorer Service from Vertex AI Workbench.

## Prerequisites
- Code Refactorer service running (locally or on Cloud Run)
- Anthropic API key configured
- Python packages: `requests`, `json`

## Setup

```python
from typing import Any

import requests

# Configure the service URL
# For local: SERVICE_URL = "http://localhost:8080"
# For Cloud Run: SERVICE_URL = "https://your-service-url.run.app"
SERVICE_URL = "http://localhost:8080"


def refactor_code(code: str, language: str, refactor_type: str = "full") -> dict[str, Any]:
    """Send code to refactoring service."""
    url = f"{SERVICE_URL}/api/v1/refactor/"
    payload = {
        "code": code,
        "language": language,
        "refactor_type": refactor_type,
        "preserve_functionality": True,
    }
    response = requests.post(url, json=payload)
    response.raise_for_status()
    return response.json()


def analyze_code(code: str, language: str) -> dict[str, Any]:
    """Analyze code quality."""
    url = f"{SERVICE_URL}/api/v1/refactor/analyze"
    payload = {"code": code, "language": language}
    response = requests.post(url, json=payload)
    response.raise_for_status()
    return response.json()


print("Setup complete! Service URL:", SERVICE_URL)
```

## Example 1: Refactor Python Code for Readability

```python
# Original code with poor readability
code = """
def f(x):
    r=[]
    for i in range(len(x)):
        if x[i]%2==0:
            r.append(x[i])
    return r
"""

result = refactor_code(code, "python", "readability")

print("Original Code:")
print(result["original_code"])
print("\n" + "=" * 50 + "\n")
print("Refactored Code:")
print(result["refactored_code"])
print("\n" + "=" * 50 + "\n")
print("Improvements:")
for improvement in result["improvements"]:
    print(f"  - {improvement}")
```

## Example 2: Optimize JavaScript Performance

```python
# JavaScript code with performance issues
js_code = """
function findDuplicates(arr) {
    var duplicates = [];
    for (var i = 0; i < arr.length; i++) {
        for (var j = i + 1; j < arr.length; j++) {
            if (arr[i] === arr[j] && duplicates.indexOf(arr[i]) === -1) {
                duplicates.push(arr[i]);
            }
        }
    }
    return duplicates;
}
"""

result = refactor_code(js_code, "javascript", "performance")

print("Refactored Code:")
print(result["refactored_code"])
print("\nComplexity Before:", result.get("complexity_before"))
print("Complexity After:", result.get("complexity_after"))
print("\nExplanation:")
print(result["explanation"])
```

## Example 3: Code Quality Analysis

```python
# Analyze code without refactoring
code_to_analyze = """
def process_data(data):
    result = []
    for item in data:
        if item > 0:
            result.append(item * 2)
    return result
"""

analysis = analyze_code(code_to_analyze, "python")

print("Quality Score:", analysis.get("overall_quality_score"))
print("\nCode Metrics:")
for metric, value in analysis["metrics"].items():
    print(f"  {metric}: {value}")

print("\nIssues Found:")
for issue in analysis["issues"]:
    print(f"  [{issue['severity'].upper()}] {issue['message']}")
    if issue.get("suggestion"):
        print(f"    Suggestion: {issue['suggestion']}")

print("\nSuggestions:")
for suggestion in analysis["suggestions"]:
    print(f"  - {suggestion}")
```

## Example 4: Full Refactoring (All Aspects)

```python
# Code with multiple issues
problematic_code = """
def calc(a,b,op):
    if op=='+':
        return a+b
    elif op=='-':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
        return a/b
"""

result = refactor_code(problematic_code, "python", "full")

print("Refactored Code:")
print(result["refactored_code"])
print("\n" + "=" * 50)
print("\nIssues Found:")
for issue in result["issues_found"]:
    print(f"  Line {issue.get('line_number', 'N/A')}: {issue['message']}")
print("\nImprovements:")
for i, improvement in enumerate(result["improvements"], 1):
    print(f"  {i}. {improvement}")
```

## Example 5: Batch Processing Multiple Files

```python
# Process multiple code snippets
code_samples = [
    ("def add(x,y): return x+y", "python"),
    ("function sum(a,b) { return a+b; }", "javascript"),
    ("fn add(x: i32, y: i32) -> i32 { x + y }", "rust"),
]

results = []
for code, lang in code_samples:
    result = refactor_code(code, lang, "best_practices")
    results.append((lang, result))

for lang, result in results:
    print(f"\n{'=' * 50}")
    print(f"Language: {lang.upper()}")
    print(f"{'=' * 50}")
    print("Refactored:")
    print(result["refactored_code"])
    print("\nKey Improvements:")
    for imp in result["improvements"][:3]:  # Show top 3
        print(f"  - {imp}")
```

## Example 6: Refactoring with Context

```python
# Provide context for better refactoring
code_with_context = """
def process(data):
    result = []
    for item in data:
        result.append(item['value'] * 1.1)
    return result
"""

url = f"{SERVICE_URL}/api/v1/refactor/"
payload = {
    "code": code_with_context,
    "language": "python",
    "refactor_type": "full",
    "context": "This function applies a 10% markup to product prices in an e-commerce system",
    "preserve_functionality": True,
}

response = requests.post(url, json=payload)
result = response.json()

print("Refactored Code (with domain context):")
print(result["refactored_code"])
print("\nExplanation:")
print(result["explanation"])
```
