## Step 5: Test pinkln Skills

```python
# Test 1: Complexity Assessment
test_problems = [
    "Calculate 2 + 2",
    "Should we build vs. buy for our CRM needs?",
    "Design a complete multi-agent system for enterprise workflow automation with real-time collaboration, version control, and intelligent task distribution.",
]

print("Complexity Assessment Test:\n")
for problem in test_problems:
    complexity = pinkln_os.assess_complexity(problem)
    strategy = pinkln_os.select_reasoning_strategy(complexity)
    print(f"Problem: {problem[:60]}...")
    print(f"  Complexity: {complexity:.2f}")
    print(f"  Strategy: {strategy}")
    print()
```

## Step 6: Create Vertex AI Integration Wrapper

```python
class VertexPnklnAgent:
    """
    Wrapper class to integrate pinkln Agent System with Vertex AI.

    This class handles:
    - Claude API calls via Vertex AI
    - pinkln OS orchestration
    - Response formatting and validation
    """

    def __init__(self, project_id: str, location: str, model: str = "claude-3-5-sonnet-v2@20241022"):
        self.client = AnthropicVertex(region=location, project_id=project_id)
        self.model = model
        self.pinkln_os = PnklnOS()

    def execute(self, challenge: str, max_tokens: int = 4096, temperature: float = 1.0) -> dict:
        """
        Execute a challenge using the pinkln OS and Claude on Vertex AI.

        Args:
            challenge: The problem to solve
            max_tokens: Maximum tokens in response
            temperature: Sampling temperature

        Returns:
            Response dictionary with solution and metadata
        """
        # Get pinkln system prompt
        system_prompt = self.pinkln_os.get_system_prompt()

        # Assess complexity and select strategy
        complexity = self.pinkln_os.assess_complexity(challenge)
        strategy = self.pinkln_os.select_reasoning_strategy(complexity)

        # Enhance the challenge with strategy guidance
        enhanced_prompt = f"""
{challenge}

Apply {strategy} reasoning to this challenge.
Remember the core directives: Question Everything, Obsess Over Details, Plan Like Da Vinci,
Craft Don't Just Code, Iterate Relentlessly, Simplify Ruthlessly.
"""

        # Call Claude via Vertex AI
        response = self.client.messages.create(
            model=self.model,
            max_tokens=max_tokens,
            temperature=temperature,
            system=system_prompt,
            messages=[{"role": "user", "content": enhanced_prompt}],
        )

        return {
            "solution": response.content[0].text,
            "complexity": complexity,
            "strategy": strategy,
            "model": self.model,
            "usage": {"input_tokens": response.usage.input_tokens, "output_tokens": response.usage.output_tokens},
        }


print("✓ VertexPnklnAgent class defined")
```

## Step 7: Run a Test Challenge

```python
# Initialize the Vertex-integrated pinkln agent
vertex_agent = VertexPnklnAgent(project_id=PROJECT_ID, location=LOCATION)

# Test with a sample challenge
test_challenge = """Design a simple revenue optimization strategy for a SaaS product 
that currently has 1000 users at $10/month. What are the top 3 opportunities?"""

print("Executing test challenge...\n")
result = vertex_agent.execute(test_challenge)

print("=" * 60)
print(f"CHALLENGE: {test_challenge}")
print("=" * 60)
print(f"\nComplexity: {result['complexity']:.2f}")
print(f"Strategy: {result['strategy']}")
print(f"\nSOLUTION:\n{result['solution']}")
print("\n" + "=" * 60)
print(f"Tokens used: {result['usage']['input_tokens']} in, {result['usage']['output_tokens']} out")
print("=" * 60)
```
